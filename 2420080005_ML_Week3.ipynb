```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA

data = pd.read_csv("/content/Titanic-Dataset (2).csv")

numeric_data = data.select_dtypes(include=[np.number])
numeric_data = numeric_data.fillna(numeric_data.mean())

scaler = StandardScaler()
data_standardized = scaler.fit_transform(numeric_data)

normalizer = MinMaxScaler()
data_normalized = normalizer.fit_transform(numeric_data)

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_standardized)

print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)

plt.figure(figsize=(6,5))
plt.scatter(data_pca[:,0], data_pca[:,1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA 2D Visualization - Titanic Dataset")
plt.show()

pca_full = PCA()
pca_full.fit(data_standardized)

cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)

plt.figure(figsize=(6,5))
plt.plot(cumulative_variance, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot - Titanic Dataset")
plt.show()
```
