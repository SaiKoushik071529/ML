```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
import networkx as nx
```

```python
transactions = [
    ['Bread', 'Milk', 'Eggs'],
    ['Bread', 'Butter', 'Jam'],
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Milk', 'Butter', 'Eggs'],
    ['Milk', 'Eggs', 'Cereal'],
    ['Bread', 'Eggs', 'Milk'],
    ['Butter', 'Jam', 'Bread'],
    ['Bread', 'Milk', 'Butter'],
    ['Milk', 'Cereal', 'Eggs'],
    ['Bread', 'Butter', 'Jam', 'Milk']
]

print("Total transactions:", len(transactions))
print("Sample transactions:", transactions[:3])
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)
print(df.head())
```

```python
item_freq = df.sum().sort_values(ascending=False)
print(item_freq)
```

```python
trans_lengths = [len(t) for t in transactions]

print("Average items:", np.mean(trans_lengths))
print("Min items:", min(trans_lengths))
print("Max items:", max(trans_lengths))
```

```python
min_support = 0.3

frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)

print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

print(rules[['antecedents','consequents','support','confidence','lift']])
```

```python
rules = rules[rules['confidence'] >= 0.5]

print("Filtered Rules:")
print(rules)
```

```python
top_lift = rules.nlargest(5, 'lift')

for i, (idx,row) in enumerate(top_lift.iterrows(),1):
    ante = ', '.join(list(row['antecedents']))
    cons = ', '.join(list(row['consequents']))

    print(f"{i}. {ante} -> {cons}")
    print("Support:", row['support'])
    print("Confidence:", row['confidence'])
    print("Lift:", row['lift'])
```

```python
top_conf = rules.nlargest(5, 'confidence')

for i, (idx,row) in enumerate(top_conf.iterrows(),1):
    ante = ', '.join(list(row['antecedents']))
    cons = ', '.join(list(row['consequents']))

    print(f"{i}. {ante} -> {cons}")
```

```python
plt.figure(figsize=(10,6))

plt.bar(range(len(item_freq)), item_freq.values)

plt.xticks(range(len(item_freq)), item_freq.index, rotation=45)

plt.title("Item Frequencies")
plt.xlabel("Items")
plt.ylabel("Frequency")

plt.show()
```

```python
G = nx.DiGraph()

for idx,row in top_lift.iterrows():
    ante = ', '.join(list(row['antecedents']))
    cons = ', '.join(list(row['consequents']))
    G.add_edge(ante, cons, weight=row['lift'])

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True,
        node_color='lightblue',
        node_size=2000,
        arrows=True)

plt.title("Association Rules Network")
plt.show()
```

```python
print("Interpretation:")

best_rule = rules.nlargest(1,'lift').iloc[0]

ante = ', '.join(list(best_rule['antecedents']))
cons = ', '.join(list(best_rule['consequents']))

print("Strongest rule:", ante, "->", cons)
print("Lift:", best_rule['lift'])
```

```python
rules[['antecedents','consequents','support','confidence','lift']].to_csv("rules.csv",index=False)

print("Rules saved to rules.csv")
```
